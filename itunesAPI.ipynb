```python
import requests
import json
import pandas as pd
base_url = "https://itunes.apple.com/search"
```

```python
r = requests.get(base_url, params = {"term":"the beatles", "country":"IN"})
r.status_code
```

```python
info = r.json()
print(json.dumps(info, indent=4))
```

```python
songs = pd.DataFrame(info["results"])
```
